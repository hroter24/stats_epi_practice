```python
import seaborn as sns
import pandas as pd
from scipy.stats import ttest_ind

# Load
df = sns.load_dataset('penguins')

# Clean
df = df.dropna(subset=['body_mass_g', 'sex'])

# Split groups
males = df[df['sex'] == 'Male']['body_mass_g']
females = df[df['sex'] == 'Female']['body_mass_g']

# Run t-test
t_stat, p_val = ttest_ind(males, females)
print("t-statistic:", t_stat)
print("p-value:", p_val)
```

```python
import numpy as np
import seaborn as sns

# Load and clean data
df = sns.load_dataset("penguins").dropna(subset=["body_mass_g", "sex"])

males = df[df["sex"]=="Male"]["body_mass_g"]
females = df[df["sex"]=="Female"]["body_mass_g"]

# Means
mean_m = males.mean()
mean_f = females.mean()

# Standard deviations
std_m = males.std()
std_f = females.std()

# Sample sizes
n_m = males.count()
n_f = females.count()

# Pooled standard deviation
s_pooled = np.sqrt(
    ((n_m - 1)*std_m**2 + (n_f - 1)*std_f**2) / (n_m + n_f - 2)
)

# Cohen's d
cohens_d = (mean_m - mean_f) / s_pooled
cohens_d
```
