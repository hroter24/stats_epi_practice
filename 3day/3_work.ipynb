```python
import pandas as pd

df = pd.DataFrame(
    {
        "Biologic": [42, 358, 400],
        "Standard": [25, 375, 400],
    },
    index=["Infection", "No Infection", "Total"]
)

df.index.name = "Outcome"
print(df)
```

```python
# calcualte the risk of infection in each group
risk_biologic = df.loc["Infection", "Biologic"] / df.loc["Total", "Biologic"]
risk_standard = df.loc["Infection", "Standard"] / df.loc["Total", "Standard"]
print(f"Risk of infection in Biologic group: {risk_biologic:.2%}")
print(f"Risk of infection in Standard group: {risk_standard:.2%}")
```

```python
risk_biologic = df.loc["Infection", "Biologic"] / df.loc["Total", "Biologic"]
risk_standard = df.loc["Infection", "Standard"] / df.loc["Total", "Standard"]

risk_difference = risk_biologic - risk_standard
print(f"Risk difference: {risk_difference:.2%}") 

risk_ratio = risk_biologic / risk_standard
print(f"Risk ratio: {risk_ratio:.2f}")

odds_biologic = df.loc["Infection", "Biologic"] / df.loc["No Infection", "Biologic"]
odds_standard = df.loc["Infection", "Standard"] / df.loc["No Infection", "Standard"]
odds_ratio = odds_biologic / odds_standard
print(f"Odds ratio: {odds_ratio:.2f}")
 
```
